```python
import opendatasets as od
dataset_url = 'https://www.kaggle.com/aaron7sun/stocknews'
od.download(dataset_url)
```

```python
import pandas as pd
df=pd.read_csv("/content/stocknews/Combined_News_DJIA.csv")
df
```

```python
df.shape
```

```python
train=df[df["Date"]<"2015-01-01"]
test=df[df["Date"]>"2014-12-31"]
```

```python
train
```

```python
from dataclasses import replace
#removing puncations

data=train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex=True,inplace=True)
```

```python
data
```

```python
#chnge the column name from top1 to 1

list1=[i for i in range(25)]
new_index=[str(i) for i in list1]
data.columns=new_index
data.head(5)
```

```python
#converting headings to lower case

for index in new_index:
  data[index]=data[index].str.lower()
data.head(5)
```

```python
#combining row in to a paragraph
''.join(str(x) for x in data.iloc[1,0:25])
```

```python
#convert all in to paragraph

headlines=[]
for row in range(0,len(data.index)):
  headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
headlines[1]
```

```python
new_strings = []

for string in headlines:
    new_string = string.replace(" b ", " ")
    new_string=new_string[1:]
    new_strings.append(new_string)

new_strings[1]
headlines=new_strings
```

```python
headlines[2]
```

```python
df = pd.DataFrame(headlines)
```

```python
df
```

```python
#download data set

df.to_csv('f.csv',index=False)
```

```python
df=headlines
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(headlines)
```

```python
traindataset
```

```python
#impliment randomforest classifier

randomclassifier=RandomForestClassifier(n_estimators=200,criterion="entropy")
randomclassifier.fit(traindataset,train['Label'])
```

```python
#prediction for the test data

test_transform= []
for row in range(0,len(test.index)):
   test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))

test_dataset=countvector.transform(test_transform)
predictions=randomclassifier.predict(test_dataset)
```

```python
predictions
```

```python
#import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test["Label"],predictions)
print(matrix)
score=accuracy_score(test["Label"],predictions)
print(score)
report=classification_report(test["Label"],predictions)
print(report)
```

# using tfid vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
tfidfvector=TfidfVectorizer(ngram_range=(2,2))
traindataset=tfidfvector.fit_transform(headlines)
```

```python
#impliment randomforest classifier

randomclassifier=RandomForestClassifier(n_estimators=200,criterion="entropy")
randomclassifier.fit(traindataset,train['Label'])
```

```python
#prediction for the test data

test_transform= []
for row in range(0,len(test.index)):
   test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))

test_dataset=tfidfvector.transform(test_transform)
predictions=randomclassifier.predict(test_dataset)
```

```python
matrix=confusion_matrix(test["Label"],predictions)
print(matrix)
score=accuracy_score(test["Label"],predictions)
print(score)
report=classification_report(test["Label"],predictions)
print(report)
```
